```python
import json
import os
```

```python
# load radar.json
with open('radar.json') as f:
    radar = json.load(f)
# 打印 radar.json，并显示中文
print(json.dumps(radar, indent=2, ensure_ascii=False))
```

```python
chirp = radar['Chirp']

# print chirp
print(json.dumps(chirp, indent=2, ensure_ascii=False))
```

```python
# 将列表里的字典转换为列表
chirp_list1 = []
for k, v in chirp[0].items():
    chirp_list1.append(v)

chirp_list2 = []
for k, v in chirp[1].items():
    chirp_list2.append(v)

print(json.dumps(chirp_list1, indent=2, ensure_ascii=False))
```

```python
radar['Chirp'] = [chirp_list1, chirp_list2]
```

```python
# save radar.json
with open('radar.json', 'w') as f:
    json.dump(radar, f, indent=2, ensure_ascii=False)
    
```
